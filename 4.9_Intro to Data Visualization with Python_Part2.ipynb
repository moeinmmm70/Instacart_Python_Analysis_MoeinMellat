### ðŸ”¹ Step 1: Setup & Load Data

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import matplotlib.pyplot as plt

# Define file path
file_path = r"C:\Users\moein\anaconda3\4 - Python Fundamentals\02 Data\Prepared Data\ords_prods_cust_merge.pkl"

# Load the dataframe
df = pd.read_pickle(file_path)
```

### ðŸ”¹ Step 2: Create One of Each Plot Type

####  Line Chart Sample

### ðŸ”¹ Step 3: Histogram â€“ order_hour_of_day

```python
# Create histogram
df['order_hour_of_day'].plot.hist(bins=24)

# Set plot title and labels
plt.title('Distribution of Orders by Hour of Day')
plt.xlabel('Hour')

# Save plot
plt.savefig(r'C:\Users\moein\anaconda3\4 - Python Fundamentals\04 Analysis\4.9\order_hour_histogram.png')

# Show plot
plt.show()
```

The histogram "**Distribution of Orders by Hour of Day**" reveals the following:

- Order activity is heavily concentrated between 9 AM and 5 PM, with a noticeable peak between 10 AM and 3 PM, where frequency reaches its maximum (~2.8 million orders).

- There are very few orders placed overnight, especially between 12 AM and 6 AM, indicating minimal customer engagement during those hours.

- After 5 PM, there's a gradual decline in order volume, showing reduced customer activity into the evening.

This distribution suggests that Instacart customers are most active during standard daytime hours, likely reflecting shopping behavior aligned with work breaks, lunch hours, or general household routines.

### ðŸ”¹ Step 6: Line Chart â€“ Age vs Number of Dependents

```python
# Create the grouped line plot
age_dependents = df.groupby('age')['n_dependants'].mean()
age_dependents.plot.line()

# Customize the plot
plt.title('Average Number of Dependents by Age')
plt.xlabel('Age')
plt.ylabel('Average Number of Dependents')

# Save plot
plt.savefig(r'C:\Users\moein\anaconda3\4 - Python Fundamentals\04 Analysis\4.9\age_vs_dependents.png')

# Show plot
plt.show()
```

### Average Number of Dependents by Age

The line chart displays the relationship between customer age and their reported number of dependents.

**Interpretation:**

- Overall, the average number of dependents remains relatively stable across age groups, generally hovering around 1.5.
- There is no strong or consistent trend indicating that either younger or older age groups have significantly more or fewer dependents.
- Minor fluctuations appear across the age spectrum, suggesting some variability, but these seem random and not indicative of a meaningful pattern.
- The chart may also reflect limitations in the data granularity or noise in customer reporting.

**Insight for Marketing:**

This suggests that **age alone might not be a strong predictor** of family size in this customer dataset. If the marketing team is targeting based on family status, they may want to incorporate additional demographic variables (like marital status or income) for better segmentation.


### ðŸ”¹ Step 7: Scatterplot â€“ Age vs Income

```python
# Create scatterplot
sns.scatterplot(x='age', y='income', data=df)

# Customize the plot
plt.title('Income Distribution by Age')
plt.xlabel('Age')
plt.ylabel('Income')

# Save plot
plt.savefig(r'C:\Users\moein\anaconda3\4 - Python Fundamentals\04 Analysis\4.9\age_vs_income.png')

# Show plot
plt.show()
```

### Income Distribution by Age

The scatterplot displays the relationship between customer age and their income levels. 

From the chart, we observe that:
- Income levels are widely dispersed across all age groups.
- There is a dense horizontal band of data points around the lower income ranges, particularly between â‚¬20,000 and â‚¬200,000, suggesting that most customers fall within this income bracket regardless of age.
- Higher income values (above â‚¬400,000) appear across various age groups but become more common among customers aged 40 and older.
- There is no clear linear correlation between age and income; income distribution appears random with respect to age, though older individuals tend to show a slightly broader income spread.

This suggests that marketing campaigns targeting high-income customers should not be based solely on age.
